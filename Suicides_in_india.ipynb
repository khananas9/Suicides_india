#### Data Pre-Processing

```python
# To check if there is any null values 

df.isnull().sum()
```

#### --> There is no null values in dataset

```python
df.State.value_counts()
```

#### > We can see values like Total (All India), Total (States), Total (Uts), Before moviong forward these values needs to be removed.


```python
# Removing unwanted Values

df=df[(df['State']!='Total (All India)') & (df['State']!='Total (States)') & (df['State']!='Total (Uts)')]
```

```python
df.State.value_counts()
```

#### Data Visualization

#### Yearwise suicide cases

```python
df.head()
causes=df[df['Type_code']=='Causes']
```

```python
causes.groupby(by='Year')['Total'].sum().index
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=causes.groupby(by='Year')['Total'].sum().index,
             y=causes.groupby(by='Year')['Total'].sum())
sns.set_style(style='darkgrid')
plt.title('Year wise number of suicide in India',fontsize=12)
plt.ylabel('Suicide Cases')

plt.show()
```

#### Statewise suicide cases

```python
#causes.head()
plt.figure(figsize=(8,6))
state=df.groupby('State',as_index=False)[['Total']].sum()
Top_state=state.sort_values(by='Total',ascending=False).head(15)
sns.barplot(data=Top_state,x='Total',y='State',palette='Reds_r')
plt.xticks(size=8)
plt.yticks(size=8)
plt.title('State wise suicide cases (Top 15) ')
plt.show()


```

####  > These are the top States where most of the sucide takes place. Maharashtra tops

```python
plt.figure(figsize=(8, 6))
state = df.groupby('State', as_index=False)[['Total']].sum()
low = state.sort_values(by='Total', ascending=True).head(5)
sns.barplot(data=low, x='State', y='Total', palette='Blues')
plt.title('States with Lowest suicides recorded')

plt.show()
```

```python
why=causes.groupby('Type',as_index=False)['Total'].sum()
top_t=why.sort_values(by='Total',ascending=False)
top_t.style.background_gradient(subset='Total',cmap='Reds')
```

```python

plt.figure(figsize=(12,10))
sns.barplot(data=top_t,x='Total',y='Type')
plt.title('Major Reasons why people commited suicide')
plt.tight_layout()
plt.show()
```

```python
age=cause.groupby(by='Age_group',as_index=False)[['Total']].sum().sort_values('Total',ascending=True)
age.style.background_gradient(subset='Total',cmap='Reds')
```

```python

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
plt.figure(figsize=(6,6))
plt.pie(x=age.Total,labels=age.Age_group,autopct='%.2f%%',colors=colors,wedgeprops={'edgecolor':'Grey'})
plt.legend(age.Age_group, title="Age Group", loc="lower left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Age Group wise suicide cases',fontweight='bold',size=15)
    
plt.axis('equal')
plt.show()
```

```python
#cause
colors=['#ff9999', '#66b3ff']
plt.figure(figsize=(6,6))
gg=cause.groupby(by='Gender',as_index=False)[['Total']].sum()
plt.pie(x=gg.Total,labels=gg.Gender,colors=colors,autopct='%.1f%%',wedgeprops={'edgecolor':'grey'})
plt.legend(gg.Gender,title='Gender',loc='lower left')
plt.title('Percentage of Suicides Over The Years By Gender')
plt.show()
```

```python
gg=cause.pivot_table(index='Year',columns='Gender',values='Total',aggfunc=sum)
gg.style.background_gradient(cmap='OrRd')
```

```python
plt.figure(figsize=(8,6))
plt.plot(gg.index,gg.Female,marker='o')
plt.plot(gg.index,gg.Male,marker='o')
plt.title('Gender wise suicide recorder per year',fontweight='bold',style='italic')
plt.legend(['Female','Male'])
plt.xlabel('Years',style='italic')
plt.ylabel('Suicide Recorded',style='italic')
for y,f in zip(gg.index,gg.Female):
    plt.text(y,f+1500,f,size=8)
for y,m in zip(gg.index,gg.Male):
    plt.text(y,m+1500,m,size=8)
plt.show()
```
